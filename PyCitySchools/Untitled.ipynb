As a first task, you've been asked to analyze the district-wide standardized test results. You'll be given access to every student's math and reading scores, as well as various information on the schools they attend. Your responsibility is to aggregate the data to and showcase obvious trends in school performance. 

```python
import json
import pandas as pd
```

```python
raw_school_df = pd.DataFrame(pd.read_csv('raw_data/schools_complete.csv'))
raw_school_df
```

```python
raw_student_df = pd.DataFrame(pd.read_csv('raw_data/students_complete.csv'))
raw_student_df.head()
```

```python
grouplabel = [0, 1]
bins = [-100,70,200]
student_df = raw_student_df.set_index('Student ID')
reading_binned = pd.cut(raw_student_df['reading_score'], bins, labels = grouplabel)
student_df['reading_pass'] = pd.to_numeric(reading_binned)
math_binned = pd.cut(raw_student_df['math_score'], bins, labels = grouplabel, )
student_df['math_pass'] = pd.to_numeric(math_binned)
student_df.head()
```

```python
student_group = student_df.groupby('school')

columns={
    #'students': student_group['name'].count(),
    'avg reading_score': round(student_group['reading_score'].mean(),2),
    'avg_math_score': round(student_group['math_score'].mean(),2),
    'reading % pass': round(student_group['reading_pass'].sum() / student_group['reading_pass'].count() * 100, 2),
    'math % pass': round(student_group['math_pass'].sum() / student_group['math_pass'].count()*100, 2)
}
student_group_df = pd.DataFrame(columns).reset_index()
student_group_df.head()
merged_df = raw_school_df.merge(student_group_df, right_on = 'school',left_on = 'name' ).set_index('School ID')

merged_df = merged_df[['school', 'type', 'size', 'budget', 'avg reading_score', 'reading % pass', 'avg_math_score', 'math % pass']]
merged_df
```
